# **Business Understanding**

## 1. Latar Belakang Bisnis
Perusahaan produk olahraga ini beroperasi di pasar yang sangat kompetitif, di mana keberhasilan ditentukan oleh kemampuan untuk mengantisipasi tren penjualan, memahami permintaan konsumen, dan menyesuaikan strategi bisnis dengan cepat. Data penjualan yang dimiliki oleh perusahaan ini mencakup informasi penting seperti harga per unit, jumlah unit terjual, total penjualan, profit operasional, dan metode penjualan. Data ini dihasilkan dari berbagai retailer dan wilayah, mencerminkan kinerja penjualan dari berbagai jenis produk olahraga.

## 2. Tujuan Bisnis
Tujuan utama dari analisis ini adalah untuk memprediksi kinerja penjualan masa depan berdasarkan data historis yang ada. Dengan prediksi yang akurat, perusahaan dapat:

- **Merencanakan inventaris:** Memastikan ketersediaan produk yang tepat di berbagai lokasi untuk menghindari kekurangan atau kelebihan stok.
- **Strategi pemasaran dan promosi:** Menargetkan kampanye pemasaran pada produk atau wilayah yang diprediksi akan memiliki permintaan tinggi.
- **Pengembangan produk baru:** Mengidentifikasi tren produk yang sedang naik daun untuk mendukung inovasi produk baru.
- **Optimasi harga:** Menyesuaikan harga produk berdasarkan prediksi permintaan untuk memaksimalkan profitabilitas.

## 3. Masalah Bisnis
Perusahaan menghadapi beberapa tantangan, termasuk:

- **Fluktuasi Permintaan:** Permintaan untuk produk olahraga bisa sangat bervariasi tergantung pada musim, tren olahraga, dan event olahraga global.
- **Pengelolaan Inventaris:** Memastikan stok produk yang cukup tanpa overstocking yang bisa menyebabkan biaya tambahan.
- **Penentuan Harga:** Menentukan harga yang kompetitif dan optimal untuk menarik konsumen sambil menjaga margin keuntungan.

## 4. Pernyataan Masalah
Perusahaan perlu membangun model prediktif yang mampu memperkirakan *Total Sales* di masa depan berdasarkan data historis seperti *Price per Unit*, *Units Sold*, dan *Operating Profit*. Model ini harus dapat memberikan panduan untuk pengambilan keputusan yang lebih baik terkait dengan pengelolaan inventaris, strategi pemasaran, dan penetapan harga.

## 5. Tujuan Analisis
Tujuan dari analisis ini adalah:

- Mengembangkan model prediktif berbasis data historis untuk memperkirakan *Total Sales* produk olahraga di masa depan.
- Mengidentifikasi variabel kunci yang mempengaruhi kinerja penjualan.
- Memberikan rekomendasi yang dapat digunakan oleh manajemen untuk meningkatkan strategi bisnis, seperti penetapan harga, promosi, dan distribusi produk.

## 6. Kriteria Kesuksesan
Keberhasilan proyek ini akan diukur melalui:

- **Akurasi Prediksi:** Model harus mampu memprediksi penjualan masa depan dengan tingkat akurasi yang tinggi.
- **Kemampuan Generalisasi:** Model harus bekerja dengan baik pada data yang tidak terlihat (*data test*), bukan hanya pada data pelatihan.
- **Implementasi Keputusan Bisnis:** Model ini diharapkan memberikan wawasan yang dapat diimplementasikan dalam keputusan bisnis nyata, seperti penetapan harga produk, perencanaan inventaris, dan pengembangan strategi pemasaran.

Dengan *Business Understanding* yang tepat, perusahaan dapat menggunakan data historisnya untuk menciptakan model prediktif yang mendukung pengambilan keputusan strategis, memastikan keunggulan kompetitif di pasar produk olahraga.

- **Cek Outliers**

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
```
